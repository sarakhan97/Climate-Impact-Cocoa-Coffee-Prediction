```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
df_2023 = pd.read_csv("arabica_clean_2023.csv")
```

```python
df_2023 = df_2023.drop(["ID", "Farm Name", "Lot Number", "Mill", "ICO Number", "Company", "Region", "Producer", "Number of Bags", "Bag Weight", "In-Country Partner", "Grading Date", "Owner", "Variety", "Status", "Color", "Expiration", "Certification Body", "Certification Address", "Certification Contact", "Quakers"], axis=1)
display(df_2023)
```

```python
df_2023["Moisture"] = df_2023["Moisture Percentage"]
display(df_2023)
```

```python
df_2023["Moisture"] = df_2023["Moisture"].astype(float) / 100
```

```python
df_2023 = df_2023.drop(["Moisture Percentage",], axis=1)
display(df_2023)
```

```python
df_2023 = pd.get_dummies(df_2023, columns=["Processing Method"], prefix="Processing Method")
```

```python
# Convert boolean values (True/False) to integers (1/0)
processing_columns = [col for col in df_2023.columns if "Processing Method_" in col]
df_2023[processing_columns] = df_2023[processing_columns].astype(int)
```

```python
display(df_2023)
```

```python
def clean_altitude(altitude):
    if pd.isnull(altitude) or altitude == "." or str(altitude).lower() == "variety":
        return np.nan

    altitude = str(altitude).replace(",", "")

    if "-" in altitude:
        numbers = [float(x) for x in altitude.split("-") if x.replace(".", "", 1).isdigit()]
        return np.mean(numbers) if numbers else np.nan

    if altitude.replace(".", "", 1).isdigit():
        return float(altitude)

    return np.nan

# Apply cleaning function
df_2023.loc[:, "Altitude"] = df_2023["Altitude"].astype(str).apply(clean_altitude)

# Fill missing values with median altitude
df_2023.loc[:, "Altitude"] = df_2023["Altitude"].fillna(df_2023["Altitude"].median())

# Display cleaned Altitude column
print(df_2023["Altitude"])
```

```python
display(df_2023)
```

```python
df_2023["Harvest Year"] = df_2023["Harvest Year"].replace("Unknown", "2023")
```

```python
def parse_harvest_year(year_str):
    """
    Convert a string like '2023' or '2023 / 2024' to a numeric value.
    For a range, we take the midpoint. E.g., '2023 / 2024' -> 2023.5
    """
    # Check if there's a slash, indicating a range
    if '/' in year_str:
        start_str, end_str = year_str.split('/')
        start_year = float(start_str.strip())
        end_year = float(end_str.strip())
        return (start_year + end_year) / 2.0
    else:
        # Single year, just parse as float
        return float(year_str.strip())

# Apply the function to the 'HarvestYear' column
df_2023['Harvest Year'] = df_2023['Harvest Year'].apply(parse_harvest_year)
```

```python
display(df_2023)
```

```python
df_2023 = df_2023.drop(["Unnamed: 0"], axis=1)
df_2023.to_csv("coffee_data_2023.csv", index=False)
```
