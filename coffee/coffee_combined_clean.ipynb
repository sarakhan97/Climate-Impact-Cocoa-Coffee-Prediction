```python
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv('cleaned_Coffee_Data.csv')
```

```python
new_order = ['Country of Origin', 'Harvest Year'] + [col for col in df.columns if col not in ['Country of Origin', 'Harvest Year']]
df = df[new_order]
```

```python
df = df.sort_values(by=['Country of Origin', 'Harvest Year'])
```

```python
df = df[df['Altitude'] >= 500]
```

```python
df.to_csv('data_reordered_clean.csv', index=False)
```

```python
df_proc = pd.read_csv('removed_outliners.csv')
```

```python
df_proc = df_proc.drop_duplicates()
```

```python
numeric_cols = df_proc.select_dtypes(include=['number']).columns
```

```python
df_proc[numeric_cols] = df_proc[numeric_cols].fillna(df_proc[numeric_cols].mean())
```

```python
df_proc.to_csv('fill_empty.csv', index=False)
```

```python
df_proc['Harvest Year'] = df_proc['Harvest Year'].astype(int)
```

```python
df_proc['Harvest Year'] = df_proc['Harvest Year'].astype(int)

# Generate unique pairs from the same DataFrame (df_proc)
unique_pairs = df_proc[['Country of Origin', 'Harvest Year']].drop_duplicates()
print("Unique pairs of Country of Origin and Harvest Year:")
print(unique_pairs)
```
