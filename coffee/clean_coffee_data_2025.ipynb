```python
import numpy as np

def clean_altitude(altitude):
    if pd.isnull(altitude) or altitude == "." or str(altitude).lower() == "variety":
        return np.nan

    altitude = str(altitude).replace(",", "")

    if "-" in altitude:
        numbers = [float(x) for x in altitude.split("-") if x.replace(".", "", 1).isdigit()]
        return np.mean(numbers) if numbers else np.nan

    if altitude.replace(".", "", 1).isdigit():
        return float(altitude)

    return np.nan

# Apply cleaning function
df_2024.loc[:, "Altitude"] = df_2024["Altitude"].astype(str).apply(clean_altitude)

# Fill missing values with median altitude
df_2024.loc[:, "Altitude"] = df_2024["Altitude"].fillna(df_2024["Altitude"].median())

# Display cleaned Altitude column
print(df_2024["Altitude"])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_columns = ["Altitude", "Aroma", "Flavor", "Acidity", "Sweetness", "Total Cup Points"]

df_2024[scaled_columns] = scaler.fit_transform(df_2024[scaled_columns])
```

```python
df_2024 = df_2024.sort_values(by=["Harvest Year_2023", "Harvest Year_2024", "Harvest Year_2025"])

# Creating lag features (previous year's data for prediction)
df_2024["Previous Aroma"] = df_2024["Aroma"].shift(1)
df_2024["Previous Flavor"] = df_2024["Flavor"].shift(1)
df_2024["Previous Acidity"] = df_2024["Acidity"].shift(1)

# Drop first row since it will have NaN values due to shifting
df_2024 = df_2024.dropna()
```

```python
display(df_2024)
```

```python
def parse_harvest_year(year_str):
    """
    Convert a string like '2023' or '2023 / 2024' to a numeric value.
    For a range, we take the midpoint. E.g., '2023 / 2024' -> 2023.5
    """
    # Check if there's a slash, indicating a range
    if '/' in year_str:
        start_str, end_str = year_str.split('/')
        start_year = float(start_str.strip())
        end_year = float(end_str.strip())
        return (start_year + end_year) / 2.0
    else:
        # Single year, just parse as float
        return float(year_str.strip())

# Apply the function to the 'HarvestYear' column
df_2024['Harvest Year'] = df_2024['Harvest Year'].apply(parse_harvest_year)
```

```python
df_2024['Category One Defects'] = df_2024['Category One Defects'].apply(lambda x: int(x.split()[0]))
df_2024['Category Two Defects'] = df_2024['Category Two Defects'].apply(lambda x: int(x.split()[0]))
```
