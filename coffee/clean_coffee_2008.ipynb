```python
import pandas as pd
from IPython.display import display
import re
import numpy as np
```

```python
df = pd.read_csv("arabica_2008.csv")
```

```python
display(df)
```

```python
df = df.drop(["Species", "Farm.Name", "Lot.Number", "Mill", "ICO.Number", "Company", "Altitude", "Region", "Producer", "Number.of.Bags", "Bag.Weight", "In.Country.Partner", "Grading.Date", "Owner.1", "Variety", "Color", "Expiration", "Certification.Body", "Certification.Address", "Certification.Contact", "altitude_low_meters", "altitude_high_meters"], axis=1)
```

```python
display(df)
```

```python
df = df.dropna(subset=['altitude_mean_meters'])
```

```python
display(df)
```

```python
df['Altitude'] = df['altitude_mean_meters'].astype(float)
```

```python
df = df.drop(['altitude_mean_meters'], axis=1)
```

```python
display(df)
```

```python
df = df.drop(['Unnamed: 0', 'Owner'], axis=1)
```

```python
display(df)
```

```python
df = df.drop(["unit_of_measurement"], axis=1)
```

```python
display(df)
```

```python
df = pd.get_dummies(df, columns=["Processing.Method"], prefix="Processing Method")
```

```python
# Convert boolean values (True/False) to integers (1/0)
processing_columns = [col for col in df.columns if "Processing Method_" in col]
df[processing_columns] = df[processing_columns].astype(int)
```

```python
display(df)
```

```python
print(df["Harvest.Year"].unique())
```

```python
def extract_year(value):
    if pd.isna(value):
        return np.nan

    # Find all years in the string
    years = re.findall(r'\d{4}', str(value))

    if len(years) == 1:
        return float(years[0])

    elif len(years) > 1:
        return float(sum(map(int, years)) / len(years))

    return np.nan

# Apply the function to the column
df["Harvest.Year"] = df["Harvest.Year"].apply(extract_year)

# Display the unique transformed values
print(df["Harvest.Year"].unique())
```

```python
print(df["Harvest.Year"].unique())
```

```python
nan_count = df["Harvest.Year"].isna().sum()
print(f"Number of NaN values in Harvest.Year: {nan_count}")
```

```python
df = df.dropna(subset=['Harvest.Year'])
```

```python
df["Harvest Year"] = df["Harvest.Year"]
```

```python
df.to_csv("coffee_data_2008.csv", index=False)
```
